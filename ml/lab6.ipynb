```python



#9.prepare X and Y

X=df.drop(columns="label")
y=df["label"]
```

```python
#10.visualization the image of the first 5 digital in X also print class label
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
for i in range(5):
    image=X.iloc[i].values.reshape(28,28)
    label=y.iloc[i]
    plt.subplot(1,5,i+1)
    plt.imshow(image,cmap="gray")
    plt.title(f"label:{label}")
    plt.axis("off")
plt.show()
```

```python
#11.split train dataset into training set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
print(X_train.shape)
```

```python

#12.fit KNN classifies to training data
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
#13.Default parameters n-neighbors=5,metric='minkowski',p=2
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)
knn.fit(X_train,y_train)
```

```python
#14.predict validation set results
y_pred=knn.predict(X_test)
print(y_pred[:5])
```

```python
#15.print confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
#16.print classification report and accuracy
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
#17.visualize image of first 5 digits in X-test also print predicted
plt.figure(figsize=(12,6))
for i in range(5):
    image=X_test.iloc[i].values.reshape(28,28)
    actual=y_test.iloc[i]
    plt.subplot(1,5,i+1)
    plt.imshow(image,cmap="gray")
    plt.title(f"predicted:{y_pred[i]},actual:{actual}")
    plt.axis("off")
plt.show()
```
