```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
train_data = pd.read_csv('train.csv')

# Set up the visualisation style
sns.set(style="whitegrid")

# Create a figure with subplots for each of the categorial variables
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

# Bar chart for Survived
sns.countplot(data=train_data, x='Survived', ax=axs[0, 0])
axs[0, 0].set_title('Number of Passengers Survived vs Not Survived')
axs[0, 0].set_xticklabels(['Did Not Survive', 'Survived'])
axs[0, 0].set_ylabel('Count')

# Bar chart for Pclass
sns.countplot(data=train_data, x='Pclass', ax=axs[0, 1],palette='pastel')
axs[0, 1].set_title('Number of Passengers by Pclass')
axs[0, 1].set_xticklabels(['1st Class', '2nd Class', '3rd Class'])
axs[0, 1].set_ylabel('Count')

# Bar chart for Sex
sns.countplot(data=train_data, x='Sex', ax=axs[1, 0])
axs[1, 0].set_title('Number of Passengers by Sex')
axs[1, 0].set_xticklabels(['Female', 'Male'])
axs[1, 0].set_ylabel('Count')

# Bar chart for Embarked
sns.countplot(data=train_data, x='Embarked', ax=axs[1, 1])
axs[1, 1].set_title('Number of Passengers by Embarked')
axs[1, 1].set_xticklabels(['C', 'Q', 'S'])  # C = Cherbourg, Q = Queenstown, S = Southampton
axs[1, 1].set_ylabel('Count')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plots
plt.show()
```

```python
#Replace the missing values in Embarked column by 'S'
df["Embarked"].fillna("S",inplace=True)
df
```

```python
df["Fare"].fillna(df["Fare"].median(),inplace=True)
df
```

```python



#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data

mean_age = df["Age"].mean()  # Computes the mean of known Age values
std_age = df["Age"].std()    # Computes the standard deviation of known Age values
print(mean_age)
print(std_age)
 #def fill_missing_age(x):
def fill_missing_age(x): 
    if pd.isnull(x):
        return np.random.normal(mean_age,std_age)
    else:
        return x
df["Age"]=df["Age"].apply(fill_missing_age)
mean_age=df["Age"].mean()
std_age=df["Age"].std()
print(mean_age)
print(std_age)
print("\nnull value after filling\n",df.isnull().sum())
    
```

```python



#Drop the columns which may not be useful in prediction of survival
#Drop PassengerId, Name, Ticket, and Cabin
df=df.drop(columns=["PassengerId","Name","Ticket","Cabin"])
print(df.head())
```

```python
#Prepare X and y
print(df.columns)
x=df.drop(columns=["Survived"])
y=df["Survived"]
```

```python
df.dtypes
```

```python
col=df.select_dtypes(include=['object'])
col
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
for c in col:
    df[c]=label_encoder.fit_transform(df[c])
df
```

```python


# Label encoding of categorical features
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()

x["Sex"]=label_encoder.fit_transform(x["Sex"])
x["Embarked"]=label_encoder.fit_transform(x["Embarked"])
x
```

```python
# concat df_categorical with original df
df=pd.concat([df,pd.get_dummies(df["Embarked"],prefix="Embarked")],axis=1)
df.drop(["Embarked"],axis=1,inplace=True)
print(df.head())
```

```python


#Do a pairplot of all features
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
sns.pairplot(df,hue="Survived")
plt.show()
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_validation=train_test_split(x,y,test_size=0.1,random_state=42,stratify=y)
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
#Print the tree

# Importing required packages for visualization
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plt.figure(figsize=(20,10))
plot_tree(model,filled=True,feature_names=x.columns, class_names=["Not Survived","Survived"])
plt.show()
```

```python
# plotting tree with max_depth=3
# Set figure size
plt.figure(figsize=(24,12))
# Plot the decision tree with max depth of 3
plot_tree(model,filled=True,feature_names=x.columns, class_names=["Not Survived","Survived"],max_depth=3)
#show the plot
plt.show()
```

```python
# Check the evaluation metrics of our default model on validation set - Confusion Matrix, Accuracy, Precision, Recall, F1 score

# Importing classification report and confusion matrix from sklearn metrics

# Making predictions

# Printing classification report
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
#making predictions on the validation set
y_pred=model.predict(x_test)

#confusion_matrix
print("confusion matrix")
print(confusion_matrix(y_validation,y_pred))

#classification Report
print("\nclassification Report:")
print(classification_report(y_validation,y_pred))

#Accuracy score
accuracy=accuracy_score(y_validation,y_pred)
print(f"\nAccuracy: {accuracy:.4f}")
```
