```python
14#create a scatterplot of 'sepallengthcm'vs'sepalwidthcm' using seaborn facecrid

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the CSV file
iris_data = pd.read_csv('Iris.csv')

# Display the first few rows of the dataset
print(iris_data.head())

# Create a FacetGrid for the scatterplot of 'sepallengthcm' vs 'sepalwidthcm'
g = sns.FacetGrid(iris_data, hue='species', height=5, aspect=1.2)
g.map(sns.scatterplot, 'sepallengthcm', 'sepalwidthcm', alpha=0.7)

# Add titles and labels
g.add_legend()
g.set_axis_labels("Sepal Length (cm)", "Sepal Width (cm)")
g.set_titles(col_template="{col_name} species")
g.fig.suptitle('Sepal Length vs Sepal Width by Species', y=1.05)

# Show the plot
plt.show()
```

```python
#29.k-fold cross validation
from sklearn.model_selection import cross_val_score,KFold
from sklearn.svm import SVC
model=SVC(C=1)
k=5
kf=KFold(n_splits=k,shuffle=True,random_state=42)
cv_scores=cross_val_score(model,x,y,cv=kf)
print(f"K-Fold cross validation with {k} splits")
print(f"Cross-validation scores: {cv_scores}")
print(f"Mean Accuracy:{np.mean(cv_scores)}")
```

```python
#30.Use grid search CV to find optimal value of c-sel decision tree lab questions
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
param_g={'C':[0.1,1,10,100,1000]}
grid=GridSearchCV(SVC(),param_g,cv=5,return_train_score=True)
grid.fit(x_train,y_train)
```

```python
best_params = grid.best_params_
best_score = grid.best_score_
print("Best Parameters:", best_params)
print("Best Cross-Validation Score: {:.2f}".format(best_score))
```

```python
best_model = grid.best_estimator_
y_pred = best_model.predict(x_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Test Set Accuracy: {:.2f}".format(accuracy_score(y_test, y_pred)))
```
